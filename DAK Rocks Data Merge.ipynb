```python
import pandas as pd
```

```python
old_df = pd.read_excel('DAK Rocks Old.xlsx')
new_df = pd.read_excel('DAK Rocks UPDATED.xlsx')
```

```python
merged_df = new_df.merge(old_df[['Claim Number', 'Include', 'Theme']], on='Claim Number', how='left')
```

```python
merged_df = pd.merge(new_df, old_df[['Claim Number', 'Include', 'Theme']], on='Claim Number', how='left', suffixes=('', '_old'))

# Update the 'Include' and 'Theme' columns in the new dataframe based on the old dataframe
merged_df['Include'] = merged_df['Include_old'].combine_first(merged_df['Include'])
merged_df['Theme'] = merged_df['Theme_old'].combine_first(merged_df['Theme'])

# Drop the temporary columns
merged_df.drop(columns=['Include_old', 'Theme_old'], inplace=True)
```
