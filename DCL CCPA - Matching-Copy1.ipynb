```python
import pandas as pd
```

```python
disney = pd.read_excel('Disney_Download (2).xlsx')
request = pd.read_csv('Guest Request (2).csv')
```

```python
disney['Full Name'] = (disney['Name First'].astype(str) + ' ' + disney['Name Last'].astype(str)).str.lower()
```

```python
request['Full Name'] = (request['FIRST NAME'].astype(str) + ' ' + request['LAST NAME'].astype(str)).str.lower()
```

```python
import pandas as pd

request['DOB'] = pd.to_datetime(request['DOB'], errors='coerce')
disney['DOB'] = pd.to_datetime(disney['DOB'], errors='coerce')

merge_name_dob = pd.merge(request, disney, on=['Full Name', 'DOB'], how='left', suffixes=('_req', '_dis'))
merge_name_email = pd.merge(request, disney, left_on=['Full Name', 'EMAIL'], right_on=['Full Name', 'Person Involved - Email'], how='left', suffixes=('_req', '_dis'))
merge_name_only = pd.merge(request, disney, on='Full Name', how='left', suffixes=('_req', '_dis'))

merge_name_dob['Match Level'] = 2  # Name and DOB match
merge_name_email['Match Level'] = 2  # Name and Email match
merge_name_only['Match Level'] = 1  # Name only match

combined_df = pd.concat([merge_name_dob, merge_name_email, merge_name_only])

print(combined_df.columns)

combined_df = combined_df.drop_duplicates(subset=['Full Name', 'DOB', 'EMAIL', 'Person Involved - Email', 'Match Level'])

matched_df = combined_df.dropna(subset=['DOB_dis', 'Person Involved - Email'])

matched_df = matched_df[['Full Name', 'EMAIL', 'DOB', 'Match Level']]

print(matched_df)
```
