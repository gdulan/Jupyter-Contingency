```python
import pandas as pd
```

```python
NonIAAPA = pd.read_excel('WDW Non Attraction Related Claims FY22-9-15-2024.xlsx')
IAAPA = pd.read_excel('WDW IAAPA Attraction Related Claims FY22-9-15-2024.xlsx')
```

```python
weather = pd.read_csv('2024-09-18 9_31am.csv')
```

```python
IAAPA['Date of Incident/Loss'] = pd.to_datetime(IAAPA['Date of Incident/Loss'], errors='coerce').dt.tz_localize(None).dt.strftime('%Y-%m-%d')
NonIAAPA['Date of Incident/Loss'] = pd.to_datetime(NonIAAPA['Date of Incident/Loss'], errors='coerce').dt.tz_localize(None).dt.strftime('%Y-%m-%d')
weather['VRSN_STRT_DTS'] = pd.to_datetime(weather['VRSN_STRT_DTS'], errors='coerce').dt.tz_localize(None).dt.strftime('%Y-%m-%d')
```

```python
MKweather = weather[weather['STATION_ID'] == 'WDW09']
ECweather = weather[weather['STATION_ID'] == 'WDW08']
AKweather = weather[weather['STATION_ID'] == 'WDW03']
DHSweather = weather[weather['STATION_ID'] == 'WDW04']
MKIAAPA = IAAPA[IAAPA['Division Desc'] == 'Magic Kingdom Park']
ECIAAPA = IAAPA[IAAPA['Division Desc'] == 'Epcot']
AKIAAPA = IAAPA[IAAPA['Division Desc'] == "Disney's Animal Kingdom Theme Park"]
DHSIAAPA = IAAPA[IAAPA['Division Desc'] == "Disney's Hollywood Studios"]
MKNON = NonIAAPA[NonIAAPA['Division Desc'] == 'Magic Kingdom Park']
ECNON = NonIAAPA[NonIAAPA['Division Desc'] == 'Epcot']
AKNON = NonIAAPA[NonIAAPA['Division Desc'] == "Disney's Animal Kingdom Theme Park"]
DHSNON = NonIAAPA[NonIAAPA['Division Desc'] == "Disney's Hollywood Studios"]
```

```python
# Merge IAAPA and MKweather data for each park
MK_merged = pd.merge(MKIAAPA, MKweather, left_on='Date of Incident/Loss', right_on='VRSN_STRT_DTS', how='inner')
EC_merged = pd.merge(ECIAAPA, ECweather, left_on='Date of Incident/Loss', right_on='VRSN_STRT_DTS', how='inner')
AK_merged = pd.merge(AKIAAPA, AKweather, left_on='Date of Incident/Loss', right_on='VRSN_STRT_DTS', how='inner')
DHS_merged = pd.merge(DHSIAAPA, DHSweather, left_on='Date of Incident/Loss', right_on='VRSN_STRT_DTS', how='inner')

# Consolidate all the merged dataframes into one
IAAPAfinal_linked_df = pd.concat([MK_merged, EC_merged, AK_merged, DHS_merged], ignore_index=True)

# Display the consolidated final dataframe
print(IAAPAfinal_linked_df)
```
