```python
import pandas as pd
```

```python
excel_file = pd.ExcelFile('Copy of FY2024 ARM Tracker.xlsx')

# Get a list of sheet names in the Excel file
sheet_names = excel_file.sheet_names

# Filter and read sheets that start with 'FY'
fy_sheets = [sheet_name for sheet_name in sheet_names if sheet_name.startswith('FY')]

# Create a dictionary to store DataFrames for each sheet
dfs = {}
for sheet_name in fy_sheets:
    dfs[sheet_name] = excel_file.parse(sheet_name)
```

```python
import pandas as pd

# Assuming you have already read and stored the DataFrames in the 'dfs' dictionary.

# Extract the fiscal year values from the sheet names
fy_values = []

for sheet_name in dfs.keys():
    if sheet_name.startswith('FY'):
        try:
            fy = int(sheet_name[2:6])  # Extract the fiscal year from the sheet name
            fy_values.append((sheet_name, fy))
        except ValueError:
            # Handle sheets with non-numeric characters in the name here
            pass

# Sort the DataFrames based on fiscal year values
sorted_fy_values = sorted(fy_values, key=lambda x: x[1])
sorted_dfs = [dfs[sheet_name] for sheet_name, _ in sorted_fy_values]

# Concatenate the DataFrames in order
stacked_df = pd.concat(sorted_dfs, axis=0, ignore_index=True)
stacked_df = stacked_df.loc[:, ~stacked_df.columns.str.contains('^Unnamed')]

```

```python
stacked_df['Request Date'] = pd.to_datetime(stacked_df['Request Date'])

# Define a function to calculate fiscal year
def calculate_fiscal_year(date):
    if date.month >= 10:
        return date.year + 1
    else:
        return date.year

# Apply the function to create the 'Fiscal Year' column
stacked_df['Fiscal Year'] = stacked_df['Request Date'].apply(calculate_fiscal_year)
```
