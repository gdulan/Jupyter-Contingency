```python
#data = pd.read_csv('Daily Payments Processed for Vendors.txt', delimiter='\t', parse_dates=['Transaction Date', 'Service Start', 'Service End'])
filtered_data = pd.read_excel('Daily Payments Processed for Vendors (95).xlsx', skiprows=3, header=0)
filtered_data = filtered_data.drop('Unnamed: 0', axis=1)
```

```python
#Filter out checks that start with 00000
filtered_data = filtered_data[~filtered_data["Check #"].str.startswith("0")]
```

```python
base_path = 'Z:\work\ACH_EOBR\Vendor_Payments\Daily Archive'
log_file_path = os.path.join(base_path, 'error_log.txt')
```

```python
def get_previous_business_day():
    result = datetime.now()
    try:
        if result.weekday() == 0:  # Monday
            result = result - timedelta(days=3)
        else:
            result = result - timedelta(days=1)
    except Exception as ex:
        result = result - timedelta(days=1)
    return result

previous_business_day = get_previous_business_day()
print(previous_business_day)
```

```python
formatted_previous_day = previous_business_day.strftime('%Y-%m-%d')
start_date = formatted_previous_day
end_date = formatted_previous_day
datestamp = previous_business_day.strftime('%m%d%Y')
```

```python
filtered_data = filtered_data[(filtered_data['Transaction Date'] >= start_date) & (filtered_data['Transaction Date'] <= end_date)]
#filtered_data = filtered_data.drop('Payee Tax ID', axis=1)
```

```python
filtered_data['Service Start'] = pd.to_datetime(filtered_data['Service Start'])
filtered_data['Service Start'] = filtered_data['Service Start'].dt.strftime('%Y-%m-%d')
filtered_data['Service End'] = pd.to_datetime(filtered_data['Service End'])
filtered_data['Service End'] = filtered_data['Service End'].dt.strftime('%Y-%m-%d')
filtered_data['Service End'] = pd.to_datetime(filtered_data['Service End'])
filtered_data['Service End'] = filtered_data['Service End'].dt.strftime('%Y-%m-%d')
filtered_data['Transaction Date'] = pd.to_datetime(filtered_data['Transaction Date'])
filtered_data['Transaction Date'] = filtered_data['Transaction Date'].dt.strftime('%Y-%m-%d')

filtered_data.drop(columns=['Payee Tax ID'], inplace=True)
filtered_data.drop(columns=['Check #'], inplace=True)
```

```python
final_order = ['Payee', 'Claim Number', 'Claimant Name', 'Transaction Date', 'Invoice Number', 'Bill Review Bill ID', 'Service Start', 'Service End', 'Amount']
filtered_data = filtered_data[final_order]
```

```python
def sanitize_filename(name):
    return name.replace('/', '_').replace('\\', '_').replace(':', '_').replace('*', '_').replace('?', '_').replace('"', '_').replace('<', '_').replace('>', '_').replace('|', '_')
```

```python
if not os.path.exists(base_path):
    os.makedirs(base_path)

date_folder = os.path.join(base_path, datestamp)
os.makedirs(date_folder, exist_ok=True)

with open(log_file_path, 'w') as log_file:
    # Group by Payee and process each group
    for payee, group in filtered_data.groupby('Payee'):
        try:
            # Sanitize the payee name
            sanitized_payee = sanitize_filename(payee)
            
            # Create folder for each vendor within the date folder
            vendor_folder = os.path.join(date_folder, sanitized_payee)
            os.makedirs(vendor_folder, exist_ok=True)

            # File path for the Excel file
            file_path = os.path.join(vendor_folder, f'{sanitized_payee}_transactions_{datestamp}.xlsx')

            # Debug: Print information about the current group
            print(f"Processing payee: {payee}")
            print(f"Saving to: {file_path}")
            print(group.head())

            # Export the group to an Excel file
            group.to_excel(file_path, index=False)

            # Debug: Confirm file has been written
            if os.path.exists(file_path):
                print(f"File {file_path} created successfully.")
            else:
                print(f"Failed to create file {file_path}.")

            # Load the workbook and select the active worksheet
            wb = load_workbook(file_path)
            ws = wb.active

            # Set header font, fill, and alignment
            header_font = Font(bold=True, color="FFFFFF")
            header_fill = PatternFill(start_color="4F81BD", end_color="4F81BD", fill_type="solid")
            header_alignment = Alignment(horizontal="center", vertical="center")

            for cell in ws[1]:
                cell.font = header_font
                cell.fill = header_fill
                cell.alignment = header_alignment

            # Adjust column widths
            for col in ws.columns:
                max_length = 0
                column = col[0].column_letter  # Get the column name
                for cell in col:
                    try:
                        if len(str(cell.value)) > max_length:
                            max_length = len(cell.value)
                    except:
                        pass
                adjusted_width = (max_length + 2)
                ws.column_dimensions[column].width = adjusted_width

            # Add total row
            total_label = 'Total'
            total_amount = group['Amount'].sum()
            last_row = ws.max_row + 1
            ws.cell(row=last_row, column=8).value = total_label
            ws.cell(row=last_row, column=9).value = total_amount  # Adjust the column index based on your "Amount" column position

            # Save the formatted workbook
            wb.save(file_path)

        except Exception as e:
            # Log the error with the payee name
            log_file.write(f"Error processing payee '{payee}': {str(e)}\n")
            print(f"Error processing payee '{payee}': {str(e)}")

print("Files have been created and formatted successfully. Check the error log for any issues.")
```
