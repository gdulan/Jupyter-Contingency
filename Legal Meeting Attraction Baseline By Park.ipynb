```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import datetime as dt
```

```python
data = pd.read_excel('LEGAL MEETING ATTRACTION RATES_VOLUME WDW.xlsx')
```

```python
def get_fiscal_year_month(date):
    year = date.year
    month = date.month
    # If the month is October, November, or December, move to the next fiscal year
    fiscal_year = year + 1 if month >= 10 else year
    # Fiscal months: October = 1, November = 2, ..., September = 12
    fiscal_month = (month - 10) % 12 + 1
    return pd.Series([fiscal_year, fiscal_month])

# Apply the function to the 'Date' column
data[['Fiscal Year', 'Fiscal Month']] = data['Date of Incident/Loss'].apply(get_fiscal_year_month)

# Step 2: Group the data by Park, Attraction, Fiscal Year, and Fiscal Month
grouped = data.groupby(['Division Desc', 'Location Code Desc', 'Fiscal Year', 'Fiscal Month']).size().reset_index(name='Incident Count')

# Step 3: (Optional) Pivot the data for easier analysis
pivoted = grouped.pivot_table(index=['Division Desc', 'Location Code Desc'],
                              columns=['Fiscal Year', 'Fiscal Month'],
                              values='Incident Count',
                              fill_value=0)

# Display a portion of the grouped data for inspection
pivoted.head()
```
